#### Posters

```python
for movie in df['IMDBid']:
#     try:
    url = f"https://imdb-api.com/en/API/Posters/{api_key}/{movie}"
    payload = {}
    headers= {}
    response = requests.request("GET", url, headers=headers, data = payload)
    imdb_api_infos[movie] = {}
    posters = json.loads(response.text.encode('utf8'))['posters']
    if len(posters) > 0:
        imdb_api_infos[movie]['poster_url'] = posters[0]['link']
    else:
        imdb_api_infos[movie]['poster_url'] = ''
#     except:
#         print(movie)
```

```python
no_poster = 0
for movie in imdb_api_infos:
    if imdb_api_infos[movie]['poster_url'] == '':
        no_poster += 1
print(no_poster)
```

#### Trailers

```python
for movie in df['IMDBid']:
#     try:
    url = f"https://imdb-api.com/en/API/Trailer/{api_key}/{movie}"
    payload = {}
    headers= {}
    response = requests.request("GET", url, headers=headers, data = payload)
    try:
        trailer = json.loads(response.text.encode('utf8'))['link']
        imdb_api_infos[movie]['trailer_url'] = trailer
    except:
        imdb_api_infos[movie]['trailer_url'] = ''
#     except:
#         print(movie)
```

```python
no_trailer = 0
for movie in imdb_api_infos:
    if imdb_api_infos[movie]['trailer_url'] == '':
        no_trailer += 1
print(no_trailer)
```

#### Rating

```python
for movie in df['IMDBid']:
#     try:
    url = f"https://imdb-api.com/en/API/Ratings/{api_key}/{movie}"
    payload = {}
    headers= {}
    response = requests.request("GET", url, headers=headers, data = payload)
    try:
        score = json.loads(response.text.encode('utf8'))['imDb']
        imdb_api_infos[movie]['IMDB'] = score
    except:
        imdb_api_infos[movie]['IMDB'] = ''
    try:
        score = json.loads(response.text.encode('utf8'))['metacritic']
        imdb_api_infos[movie]['MetaCritic'] = score
    except:
        imdb_api_infos[movie]['MetaCritic'] = ''
    try:
        score = json.loads(response.text.encode('utf8'))['theMovieDb']
        imdb_api_infos[movie]['theMovieDb'] = score
    except:
        imdb_api_infos[movie]['theMovieDb'] = ''
    try:
        score = json.loads(response.text.encode('utf8'))['rottenTomatoes']
        imdb_api_infos[movie]['RottenTomatoes'] = score
    except:
        imdb_api_infos[movie]['RottenTomatoes'] = ''
    try:
        score = json.loads(response.text.encode('utf8'))['filmAffinity']
        imdb_api_infos[movie]['FilmAffinity'] = score
    except:
        imdb_api_infos[movie]['FilmAffinity'] = ''
#     break
```

```python
no_IMDB = 0
no_MetaCritic = 0
no_theMovieDb = 0
no_RottenTomatoes = 0
no_FilmAffinity = 0

for movie in imdb_api_infos:
    if imdb_api_infos[movie]['IMDB'] == '':
        no_IMDB += 1
    if imdb_api_infos[movie]['MetaCritic'] == '':
        no_MetaCritic += 1
    if imdb_api_infos[movie]['theMovieDb'] == '':
        no_theMovieDb += 1
    if imdb_api_infos[movie]['RottenTomatoes'] == '':
        no_RottenTomatoes += 1
    if imdb_api_infos[movie]['FilmAffinity'] == '':
        no_FilmAffinity += 1
print(no_IMDB,no_MetaCritic,no_theMovieDb,no_RottenTomatoes,no_FilmAffinity)
```

#### UserRatings

```python
for movie in df['IMDBid']:
#     try:
    url = f"https://imdb-api.com/en/API/UserRatings/{api_key}/{movie}"
    payload = {}
    headers= {}
    response = requests.request("GET", url, headers=headers, data = payload)
    try:
        score = json.loads(response.text.encode('utf8'))['demographicMales']['allAges']['rating']
        imdb_api_infos[movie]['IMDB_male'] = score
    except:
        imdb_api_infos[movie]['IMDB_male'] = ''
    try:
        score = json.loads(response.text.encode('utf8'))['demographicFemales']['allAges']['rating']
        imdb_api_infos[movie]['IMDB_female'] = score
    except:
        imdb_api_infos[movie]['IMDB_female'] = ''
```

```python
no_male = 0
no_female = 0

for movie in imdb_api_infos:
    if imdb_api_infos[movie]['IMDB_male'] == '':
        no_male += 1
    if imdb_api_infos[movie]['IMDB_female'] == '':
        no_female += 1
print(no_male, no_female)
```

```python
cnt = 0
for movie in df['IMDBid']:
    try:
        url = "https://www.imdb.com/title/"+movie+"/reviews/_ajax?ref_=undefined&paginationKey={}"
        key = "g4wp7bbfqqydoyyh7ovxrmbvrptm6cjhzfmxvlnomwklyczuf43o6ss7pe4vrprndr4k5xbfpgn3rldoq3jzj74iui53agq"
        data = {"title": [], "user":[], "rating":[], "date":[], 'spoil':[], "review": []}

        while True:
            response = requests.get(url.format(key))
            soup = BeautifulSoup(response.content, "html.parser")
            # Find the pagination key
            pagination_key = soup.find("div", class_="load-more-data")
            if not pagination_key:
                break
            # Update the `key` variable in-order to scrape more reviews
            key = pagination_key["data-key"]
            for title, review, date, user, content in zip(
                soup.find_all(class_="title"), 
                soup.find_all(class_="text show-more__control"), 
                soup.find_all(class_="review-date"),
                soup.find_all(class_="display-name-link"), 
                soup.findAll(class_='lister-item-content')
            ):
                data["title"].append(title.get_text(strip=True))
                data["review"].append(review.get_text())

                data["date"].append(parse(date.get_text()).strftime("%Y-%m-%d"))
                data["user"].append(user.find('a')['href'].split('/')[-2])
                if content.find(class_='spoiler-warning')!=None:
                    data["spoil"].append(1)
                else:
                    data["spoil"].append(0)
                if content.find(class_='rating-other-user-rating')!=None:
                    data["rating"].append(content.find(class_='rating-other-user-rating').get_text(strip=True).split('/')[0])
                else:
                    data["rating"].append(None)
        pd.DataFrame(data).to_csv('reviews_rat/'+movie+'.csv', index=False)
    except:
        print(movie)
    print(cnt, end=' ')
    cnt += 1
```
